display(HTML("""
<p style="color:#0B8043; font-size:22px; font-weight:bold;">
  Modelo Matemático
</p>

<p style="font-size:16px;">
Sea Xₙ la posición después de n pasos.
En cada paso:
</p>

<p style="font-size:18px;">
$$
\\Delta X =
\\begin{cases}
+1 & con\\ probabilidad\\ p \\\\
-1 & con\\ probabilidad\\ 1-p
\\end{cases}
$$
</p>

<p style="font-size:16px;">
La posición total es:
</p>

<p style="font-size:18px;">
$$
X_n = \\sum_{i=1}^{n} \\Delta X_i
$$
</p>

<p style="font-size:16px;">
Donde p + q = 1.
</p>
"""))


```python
import random as r
import matplotlib.pyplot as plt
```

```python
# ====== PARÁMETROS ======

p = 0.5      # Probabilidad de ir a la derecha
n = 200      # Número de pasos
```

```python
posicion = 0
trayectoria = [0]

# Caminata
for i in range(n):

    numero = r.random()   # Número entre 0 y 1

    if numero < p:
        posicion += 1     # Derecha
    else:
        posicion -= 1     # Izquierda

    trayectoria.append(posicion)
```

```python
# Gráfica
plt.figure(figsize=(9,5))
plt.plot(trayectoria)
plt.axhline(0)
plt.title("Caminata Aleatoria 1D")
plt.xlabel("Pasos")
plt.ylabel("Posición")
plt.grid(True)
plt.show()

print("Posición final:", posicion)
```

```python
finales = []

repeticiones = 1000   # Cuántas caminatas simular

for j in range(repeticiones):

    pos = 0

    for i in range(n):

        if r.random() < p:
            pos += 1
        else:
            pos -= 1

    finales.append(pos)

plt.hist(finales, bins=30)
plt.title("Distribución de posiciones finales")
plt.xlabel("Posición final")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()
```
