Enteros: Representan números enteros, positivos o negativos. Por ejemplo, 10, 0 y -5 son enteros.

Flotantes: Representan números decimales. Por ejemplo, 3.14, 0.0 y -2.5 son flotantes.

Cadenas: Representan una serie de caracteres. Por ejemplo, ‘Hola mundo’ y ‘1234’ son cadenas.

Booleanos: Representan valores verdadero o falso. El valor verdadero se representa con True y el valor falso con False.
